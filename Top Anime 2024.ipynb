## Finding the rows where Episode is Unknown and Replacing them

```python

unknown_rows = data[data['Episodes'] == 'Unknown']
```

```python
print(unknown_rows)
```

```python
# Values to replace 'Unknown' with, one for each 'Unknown'
replacement_values = [1111, 267, 1132]

# Find the indices of rows where 'Episodes' is 'Unknown'
unknown_indices = data[data['Episodes'] == 'Unknown'].index

# Replace each 'Unknown' value one by one with the corresponding value from replacement_values
for i, index in enumerate(unknown_indices):
    data.loc[index, 'Episodes'] = replacement_values[i]
```

# Anime Analysis


## Top 100 Scoring Anime

```python
top_scored = data.sort_values(by=["Score"], ascending=False)
top_scored = top_scored.head(100)  
print(top_scored['English'])
```

```python
fig = px.bar(top_scored, 
             x='English',  # x-axis: English titles
             y='Score',    # y-axis: Scores
             title="Top 100 Scored Anime",  # Chart title
             width=1350,   # Set the width of the chart
             height=800)   # Set the height of the chart

fig.update_xaxes(tickangle=45)

fig.show()
```

## Top 100 Polular Anime 

```python
top_popular = data.sort_values(by=["Popularity"], ascending = False)
top_popular = top_popular.head(100)
print(top_popular['English'])
```

```python
fig = px.bar(top_popular,
             x='English',
             y='Popularity',
             title='Top 100 Popular Anime',
             width= 1350,
             height= 800)

fig.update_xaxes(tickangle=45)
fig.show()
```

## Top 100 Ranking Anime

```python
top_ranked = data.sort_values(by=['Rank'],ascending = False)
top_ranked = top_ranked.head(100)
print(top_ranked['English'])
```

```python
fig = px.bar(top_ranked,
             x='English',
             y='Rank',
             title='Top 100 Ranked Anime',
             width=1350,
             height=800)

fig.update_xaxes(tickangle=45)
fig.show()
```

## Relationship Between Duration and Rank

```python
heatmap_data = data.groupby(['Duration', 'Rank']).size().reset_index(name='count')

# Create the heatmap using Plotly Express
fig = px.density_heatmap(heatmap_data,
                         x='Duration',
                         y='Rank',
                         z='count',
                         title='Heatmap of Duration and Ranking',
                         color_continuous_scale='Spectral',
                         width=800,
                         height=800)

# Rotate x-axis labels by 45 degrees
fig.update_xaxes(tickangle=45)

# Show the chart
fig.show()
```

```python
fig = px.scatter(data, 
                 x='Duration', 
                 y='Rank', 
                 size='Score', 
                 title='Relationship Between Duration and Rank', 
                 labels={'Duration': 'Duration (min)', 'Rank': 'Rank'}, 
                 hover_data=['Score'])
fig.update_layout(
    width=1350,  # Set the width of the chart
    height=800   # Set the height of the chart
)
# Show the chart
fig.show()
```

## Relationship Between Duration and Popularity

```python
fig = px.scatter(data, 
                 x='Duration', 
                 y='Popularity', 
                 size='Score', 
                 title='Relationship Between Duration and Popularity', 
                 labels={'Duration': 'Duration (min)', 'Popularity': 'Popularity'}, 
                 hover_data=['Score'])
fig.update_layout(
    width=1350,  # Set the width of the chart
    height=800   # Set the height of the chart
)
# Show the chart
fig.show()
```

## Relationship Between Score and Popularity

```python
heatmap_data = data.groupby(['Popularity', 'Score']).size().reset_index(name='Count')

# Create the heatmap using Plotly Express
fig = px.density_heatmap(heatmap_data,
                         x='Popularity',
                         y='Score',
                         z='Count',
                         title='Heatmap of Score and Popularity',
                         color_continuous_scale='Twilight',
                         width=800,
                         height=800)

# Rotate x-axis labels by 45 degrees
fig.update_xaxes(tickangle=45)

# Show the chart
fig.show()
```
